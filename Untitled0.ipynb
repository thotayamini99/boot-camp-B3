```python
#step3:handling missing data
#scenario 1:missing value in numerical column
#scenario 2:missing values in non numerical column
#scenario 3:having missing values in huge data
import numpy as np
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x[:,1:3])
x[:,1:3]=imputer.transform(x[:,1:3])

```

```python
x
```

```python
#step 4 : handling catogorial data
#which converts the text data into numerical data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
#Alexa-0,0,1
ct=ColumnTransformer(transformers=[('enconder',OneHotEncoder(),[0])],remainder='passthrough')#encoder is a method which coverts text to numerical value
x=ct.fit_transform(x)
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
#step5:split the data into training and testing data
from sklearn.model_selection import train_test_split # train test is a class which as imported from model selecection
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#step6:feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train[:,3:]=sc.fit_transform(x_train[:,3:])
x_test[:,3:]=sc.transform(x_test[:,3:])
```
